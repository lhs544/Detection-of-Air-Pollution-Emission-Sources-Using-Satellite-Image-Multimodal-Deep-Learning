# 8주차 Multimodal Fusion (Image × Image) – 전체 개요와 단계별 설명

## 목표
- **CT + MRI** 두 영상 모달리티의 정보를 융합해 **뇌종양 진단** 성능과 영상 품질을 동시에 향상  
- **세 수준의 융합** 시연  
  1. **Feature‑level** : CNN 기반 특징추출 → 특징 결합 → 전통 ML 분류  
  2. **Pixel‑level** : 원본 영상 정합 후 평균, ECC, SIFT 기반 융합 비교  
  3. **Decision‑level** : 다중 분류기(Random Forest, MLP) 성능 비교  
- 정보 보존(Entropy·MI), 노이즈 억제(PSNR·MSE), 분류 정확도 등을 종합 평가

```python
# ===============================================================
# 8주차 Image×Image 멀티모달 융합 실습
# 각 셀은 ‘Step N is running…’ 로그를 기준으로 나뉨
# ===============================================================

# ---------------------------------------------------------------
# Step 1  ▸ 라이브러리 및 환경 설정
# ---------------------------------------------------------------
print("Step 1 is running...")

import os, cv2, numpy as np, matplotlib.pyplot as plt

# 딥러닝 & 모델 관련
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D,
                                     Flatten, Dense, Dropout)
from tensorflow.keras.applications import VGG19
from tensorflow.keras.applications.vgg19 import preprocess_input

# 학습/평가 유틸
from sklearn.model_selection import train_test_split

print("Step 1 completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 2  ▸ CT/MRI 데이터 로드 & 전처리
# ---------------------------------------------------------------
print("Step 2 is running...")

def load_images_from_folder_limited(folder, label, limit=50, target_size=(128,128)):
    """
    지정된 폴더에서 최대 `limit` 장까지 흑백 이미지를 읽어
    ▸ 128×128 리사이즈
    ▸ 그레이→RGB 변환(채널 3개 맞춤)
    ▸ 0~1 정규화
    그리고 라벨 배열과 함께 반환
    """
    images, labels, count = [], [], 0
    for filename in sorted(os.listdir(folder)):
        if count >= limit:          # 개수 제한
            break
        img_path = os.path.join(folder, filename)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, target_size)
            img_rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB) / 255.0
            images.append(img_rgb)
            labels.append(label)
            count += 1
    return np.array(images), np.array(labels)

# ▸ CT 이미지(종양=1, 정상=0) -----------------------------------
print("Loading CT images...")
x_ct_tumor,   y_ct_tumor   = load_images_from_folder_limited(
    "/kaggle/input/brain-tumor-multimodal-image-ct-and-mri/Dataset/Brain Tumor CT scan Images/Tumor",
    label=1, limit=50, target_size=(128,128))
x_ct_healthy, y_ct_healthy = load_images_from_folder_limited(
    "/kaggle/input/brain-tumor-multimodal-image-ct-and-mri/Dataset/Brain Tumor CT scan Images/Healthy",
    label=0, limit=50, target_size=(128,128))

# ▸ MRI 이미지(종양=1, 정상=0) ----------------------------------
print("Loading MRI images...")
x_mri_tumor,   y_mri_tumor   = load_images_from_folder_limited(
    "/kaggle/input/brain-tumor-multimodal-image-ct-and-mri/Dataset/Brain Tumor MRI images/Tumor",
    label=1, limit=50, target_size=(128,128))
x_mri_healthy, y_mri_healthy = load_images_from_folder_limited(
    "/kaggle/input/brain-tumor-multimodal-image-ct-and-mri/Dataset/Brain Tumor MRI images/Healthy",
    label=0, limit=50, target_size=(128,128))

# ▸ 네 모달(CT-Tumor/Healthy, MRI-Tumor/Healthy)을 하나로 합침 ----
X = np.concatenate([x_ct_tumor, x_ct_healthy, x_mri_tumor, x_mri_healthy], axis=0)
y = np.concatenate([y_ct_tumor, y_ct_healthy, y_mri_tumor, y_mri_healthy], axis=0)

print("Loaded images.")
print("Dataset shape:", X.shape, y.shape)
print("Step 2 completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 3  ▸ 학습/검증 세트 분할
# ---------------------------------------------------------------
print("Step 3 is running...")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)

print("Train shape:", X_train.shape)
print("Test shape :", X_test.shape)
print("Step 3 completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 4A ▸ Custom CNN 정의
# ---------------------------------------------------------------
print("Step 4A is running...")

inputs_custom = Input(shape=(128,128,3))
x = Conv2D(32, (3,3), activation='relu')(inputs_custom)
x = MaxPooling2D(2,2)(x)
x = Conv2D(64, (3,3), activation='relu')(x)
x = MaxPooling2D(2,2)(x)
x = Flatten()(x)
x = Dense(128, activation='relu', name='custom_dense')(x)  # 특징 추출층
x = Dropout(0.5)(x)
outputs_custom = Dense(1, activation='sigmoid')(x)         # 0/1 이진 분류

custom_cnn = Model(inputs_custom, outputs_custom)
custom_cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
custom_cnn.summary()

print("Step 4A completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 4B ▸ VGG16 전이학습 기반 Feature Extractor
# ---------------------------------------------------------------
print("Step 4B is running...")

from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import GlobalAveragePooling2D

vgg_base = VGG16(include_top=False, weights='imagenet', input_shape=(128,128,3))
for layer in vgg_base.layers:          # 사전학습된 가중치는 동결
    layer.trainable = False

vgg_output = GlobalAveragePooling2D()(vgg_base.output)  # Conv feature → 1D
vgg_model  = Model(inputs=vgg_base.input, outputs=vgg_output)
vgg_model.summary()

print("Step 4B completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 5  ▸ Custom CNN 학습
# ---------------------------------------------------------------
print("Step 5 is running...")

history = custom_cnn.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=5,
    batch_size=8
)

print("Step 5 completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 6  ▸ 두 모델에서 Feature 추출
# ---------------------------------------------------------------
print("Step 6 is running...")

num_samples = 50  # 시범용으로 50장만 사용

# VGG용 전처리
X_vgg = np.array([preprocess_input(x) for x in X])

# Custom CNN의 Dense층 출력 가져오기
feature_custom_model = Model(inputs=custom_cnn.input,
                             outputs=custom_cnn.get_layer('custom_dense').output)

custom_features = feature_custom_model.predict(X[:num_samples])
vgg_features    = vgg_model.predict(X_vgg[:num_samples])

print("Custom features shape:", custom_features.shape)
print("VGG16  features shape:", vgg_features.shape)
print("Step 6 completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 7  ▸ Feature‑level Fusion (Concat)
# ---------------------------------------------------------------
print("Step 7 is running...")

fused_features = np.concatenate([custom_features, vgg_features], axis=1)
print("Fused features shape:", fused_features.shape)
print("Step 7 completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 8  ▸ 정보량·유사도 지표 계산
# ---------------------------------------------------------------
print("Step 8 is running...")

from scipy.stats import entropy
from sklearn.metrics import mean_squared_error

entropies, stds, mses = [], [], []

for cf, vf, fused in zip(custom_features, vgg_features, fused_features):
    fused_norm = fused / (np.sum(fused) + 1e-8)   # PDF로 정규화
    entropies.append(entropy(fused_norm))         # Shannon Entropy
    stds.append(np.std(fused))                    # 분산(질적 다양성)
    mses.append(mean_squared_error(cf, vf[:cf.shape[0]]))  # 두 벡터 차이

print(f"Mean Entropy : {np.mean(entropies):.4f}")
print(f"Mean Std Dev : {np.mean(stds):.4f}")
print(f"Mean MSE(custom vs VGG16) : {np.mean(mses):.4f}")
print("Step 8 completed. Fusion evaluation done.")


# ---------------------------------------------------------------
# Step 9  ▸ Feature 정규화(StandardScaler)
# ---------------------------------------------------------------
print("Step 9 is running...")

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
fused_features_normalized = scaler.fit_transform(fused_features)

print("Normalized fused features shape:", fused_features_normalized.shape)
print("Step 9 completed. Ready for next step.")


# ---------------------------------------------------------------
# Step 10 ▸ RandomForest & MLP 분류 성능 비교
# ---------------------------------------------------------------
print("Step 10 is running...")

y_subset = y[:num_samples]   # 50개 샘플에 대한 라벨

X_tr, X_te, y_tr, y_te = train_test_split(
    fused_features_normalized, y_subset,
    test_size=0.2, random_state=42, stratify=y_subset)

# ▸ Random Forest ------------------------------------------------
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_tr, y_tr)
rf_preds = rf.predict(X_te)

print("Random Forest Classifier:")
from sklearn.metrics import classification_report
print(classification_report(y_te, rf_preds))

# ▸ MLP ----------------------------------------------------------
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(128,), max_iter=500, random_state=42)
mlp.fit(X_tr, y_tr)
mlp_preds = mlp.predict(X_te)

print("MLP Classifier:")
print(classification_report(y_te, mlp_preds))

print("Step 10 completed. Classifiers trained and evaluated.")


# ---------------------------------------------------------------
# Step 11 ▸ 단일 샘플에 대해 Feature 시각화 + Pixel‑level Fusion
# ---------------------------------------------------------------
print("Step 11 (Single Image Fusion) is running...")

# ---------- (1) Feature Map 기반 시각화 -------------------------
from tensorflow.keras.preprocessing.image import img_to_array, load_img

def preprocess_single_image(image_path):
    """단일 흑백 이미지를 128×128 RGB & 0~1 스케일로 변환"""
    img = load_img(image_path, color_mode='grayscale', target_size=(128, 128))
    img = img_to_array(img) / 255.0
    img_rgb = np.repeat(img, 3, axis=2)           # 1채널→3채널
    return np.expand_dims(img_rgb, axis=0)

ct_path  = "/kaggle/input/test-dataset/ct.jpg"
mri_path = "/kaggle/input/test-dataset/mri.jpg"

ct_img  = preprocess_single_image(ct_path)
mri_img = preprocess_single_image(mri_path)

# 인퍼런스 한 번 호출해 그래프 생성
_ = custom_cnn(ct_img)

# Custom CNN의 Dense 앞 출력 추출
feature_extractor = Model(inputs=custom_cnn.input,
                          outputs=custom_cnn.layers[-3].output)

ct_feat  = feature_extractor.predict(ct_img)
mri_feat = feature_extractor.predict(mri_img)

fused_feat = np.concatenate([ct_feat, mri_feat], axis=-1)

# 256‑D → 16×16 그리드 재배열 후 시각화
from sklearn.preprocessing import StandardScaler
fused_scaled = StandardScaler().fit_transform(fused_feat)
fused_image  = fused_scaled.reshape(16, 16)

plt.figure(figsize=(4,4))
plt.imshow(fused_image, cmap="viridis")
plt.title("Fused Feature Image (CT + MRI)")
plt.axis("off")
plt.show()

# ---------- (2) ECC 기반 Affine 정합 & 평균 융합 ---------------
import cv2

# 원본 그레이스케일 읽기
ct_raw  = cv2.imread(ct_path,  cv2.IMREAD_GRAYSCALE)
mri_raw = cv2.imread(mri_path, cv2.IMREAD_GRAYSCALE)

# MRI를 CT 크기에 맞춰 1차 리사이즈
mri_resized = cv2.resize(mri_raw, (ct_raw.shape[1], ct_raw.shape[0]))

# 0~255 정규화 함수
normalize = lambda img: cv2.normalize(img.astype(np.float32), None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
ct_norm  = normalize(ct_raw)
mri_norm = normalize(mri_resized)

# ECC(Affine) 정합 ← 변환 행렬 초기값(2×3 Eye)
warp_mode, warp_matrix = cv2.MOTION_AFFINE, np.eye(2,3,dtype=np.float32)
criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 5000, 1e-6)
_, warp_matrix = cv2.findTransformECC(ct_norm, mri_norm, warp_matrix, warp_mode, criteria)

# MRI 정합
mri_aligned = cv2.warpAffine(mri_norm, warp_matrix,
                             (ct_norm.shape[1], ct_norm.shape[0]),
                             flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)

# 픽셀 평균 융합 + 샤픈
fused = cv2.addWeighted(ct_norm, 0.5, mri_aligned, 0.5, 0)
sharpen_kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
fused_sharp = cv2.filter2D(fused, -1, sharpen_kernel)

# 시각화
titles = ['CT', 'MRI (Resized)', 'MRI (Aligned)', 'Fused (Sharpened)']
imgs   = [ct_norm, mri_norm, mri_aligned, fused_sharp]

plt.figure(figsize=(16,5))
for i, (img, title) in enumerate(zip(imgs, titles), 1):
    plt.subplot(1,4,i)
    plt.imshow(img, cmap='gray')
    plt.title(title); plt.axis('off')
plt.tight_layout(); plt.show()


# ---------- (3) SIFT 기반 특징 정합 & 융합 ----------------------
print(" SIFT-based CT–MRI Registration")

# SIFT 키포인트 탐지
sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(ct_raw,  None)
kp2, des2 = sift.detectAndCompute(mri_raw, None)

# BF 매칭(L2) + 정렬
bf      = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
matches = sorted(bf.match(des1, des2), key=lambda x: x.distance)

# 상위 50개 매칭 시각화용
img_matches = cv2.drawMatches(ct_raw, kp1, mri_raw, kp2, matches[:50], None, flags=2)

# 매칭 좌표 추출 → Affine 변환(RANSAC)
pts1 = np.float32([kp1[m.queryIdx].pt for m in matches]).reshape(-1,1,2)
pts2 = np.float32([kp2[m.trainIdx].pt for m in matches]).reshape(-1,1,2)
M, _ = cv2.estimateAffinePartial2D(pts2, pts1, method=cv2.RANSAC)

# MRI 정합 & 평균 융합
h, w      = ct_raw.shape
mri_align = cv2.warpAffine(mri_raw, M, (w, h))
fused_sift = cv2.addWeighted(ct_raw, 0.5, mri_align, 0.5, 0)

# 결과 시각화
plt.figure(figsize=(16,6))
for idx, (img, t) in enumerate([ (ct_raw,'CT'),
                                 (mri_raw,'MRI (Original)'),
                                 (mri_align,'MRI (Aligned)'),
                                 (fused_sift,'Fused Image') ], 1):
    plt.subplot(1,4,idx); plt.imshow(img, cmap='gray'); plt.title(t); plt.axis('off')
plt.tight_layout(); plt.show()

# 매칭 시각화
plt.figure(figsize=(16,8))
plt.imshow(img_matches); plt.title('Top Feature Matches'); plt.axis('off'); plt.show()


# ---------- (4) 통합: 키포인트·호모그래피·융합 과정 세부 시각화 --
print("🟢 Integrated Feature Extraction + SIFT-based CT–MRI Registration")

# (기존 코드와 동일 – 중복 부분 생략 가능, 교육용 디테일 시각화 셀)

# …(생략)…

print("Step 11 completed. Fused image generated.")
```
